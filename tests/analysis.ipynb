```python
import numpy as np
import pandas as pd
```

```python
def parse_results(path):
    data = pd.read_csv(path)
    times = data.time.dropna()
    return pd.DataFrame(
        {
            'time': times.sort_values().reset_index(drop=True),
            'solved': np.arange(1, times.size + 1)
        }
    )
```

```python
results = parse_results('results.csv')
# results_new = parse_results('results-new.csv')

for logx in [False, True]:
    ax = results.plot(x='time', y='solved', logx=logx)
    # results_new.plot(x='time', y='solved', logx=logx, ax=ax)
```
